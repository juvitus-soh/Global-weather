```python
import matplotlib.pyplot as plt
import seaborn as sns

#Time Series Analysis of Temperature over time
data['last_updated'] = pd.to_datetime(data['last_updated'])
plt.figure(figsize=(24, 12))
sns.lineplot(x='last_updated', y='temperature_celsius', data=data, ci=None)
plt.title('Temperature Over Time')
plt.xlabel('Time')
plt.ylabel('Temperature (Celsius)')
plt.show()
```

```python
#correlation analysis
correlation_matrix = data.corr()
plt.figure(figsize=(24, 16))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
#Wind Speed Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['wind_kph'], bins=20, kde=True)
plt.title('Wind Speed Distribution')
plt.xlabel('Wind Speed (kph)')
plt.ylabel('Frequency')
plt.show()
```

```python
#scatter plot of geospatial analysis
plt.figure(figsize=(10, 8))
sns.scatterplot(x='longitude', y='latitude', hue='temperature_celsius', size='temperature_celsius', data=data)
plt.title('Geospatial Analysis of Temperature')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
# Scatter plot of temperatures against global locations
plt.figure(figsize=(12, 8))
sns.scatterplot(x='longitude', y='latitude', hue='temperature_celsius', size='temperature_celsius', data=data, palette='coolwarm', alpha=0.7)
plt.title('Temperature Across Global Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Temperature (Â°C)')
plt.show()
```
